*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Assuming spacex_df is your DataFrame
# Define a function to determine the marker color based on the class value
def get_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'

# Apply the function to the 'class' column and create a new 'marker_color' column
spacex_df['marker_color'] = spacex_df['class'].apply(get_marker_color)
```

```python
#To add a folium.Marker for each launch result in the spacex_df DataFrame to the marker_cluster, you can follow these steps:

#We will ensure that the marker_cluster is added to our site_map.


from folium.plugins import MarkerCluster

# Initializing the marker cluster
marker_cluster = MarkerCluster().add_to(site_map)
#Adding Markers to the Marker Cluster: we loop through the spacex_df DataFrame and create a folium.Marker for each row, using the coordinates and the marker_color we defined earlier.


for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate and customize the Marker's icon property
    marker = folium.Marker(
        location=(record['Lat'], record['Long']),
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"Launch Site: {record['Launch Site']}, Class: {record['class']}"
    )
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)
#Display the Map: Finally, display the map to see the markers added to the cluster.


site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
